```python
y, sr= librosa.load('/content/drive/MyDrive/fff/folddd1/fold1/15564-2-0-1.wav', sr=32000)
```

```python
mel_bins = 64
fmin = 0
fmax= None
Mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, hop_length=hop_length,window=window_type, n_mels = mel_bins, power=2.0)
print("The shape of mel spectrogram is: ",filename)

librosa.display.specshow(Mel_spectrogram, sr=sr, x_axis='time', y_axis='mel',hop_length=hop_length)
plt.colorbar(format='%+2.0f dB')
plt.title('City Voice2')
plt.tight_layout()
plt.show()
```

```python
mel_spectrogram_db = librosa.power_to_db(Mel_spectrogram, ref=np.max)
print("The shape of Log Mel spectrogram is: ", mel_spectrogram_db.shape)
librosa.display.specshow(mel_spectrogram_db, sr=sr, x_axis='time', y_axis='mel',hop_length=hop_length)
plt.colorbar(format='%+2.0f dB')
plt.title('City Voice')
plt.tight_layout()
plt.show()
```
